```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
einstein=cv2.imread('einstein.jpg',0)
mother_teresa=cv2.imread('mother_teresa.jpg',0)
solvay=cv2.imread('solvay_conference.jpg',0)
```

```python
plt.subplot(1,3,1)
plt.imshow(einstein,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(mother_teresa,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(solvay,cmap='gray')
plt.tight_layout()
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face_img=einstein.copy()
face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)

for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),2)
plt.imshow(face_img,cmap='gray')
    
```

```python
def detect(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,,scaleFactor=1.2,minNeighbors=2)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),2)
    plt.imshow(face_img,cmap='gray')
```

```python
detect(mother_teresa)
```

```python
detect(solvay)
```

```python
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
eye_img=mother_teresa.copy()
eye_rects=eye_cascade.detectMultiScale(eye_img,scaleFactor=1.2,minNeighbors=2)
for (x,y,w,h) in eye_rects:
    cv2.rectangle(eye_img,(x,y),(x+w,y+h),(255,255,255),2)
plt.imshow(eye_img,cmap='gray')
```
